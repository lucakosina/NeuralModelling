```python
import numpy as np
import matplotlib.pyplot as plt

# Ensure the data is in float format and drop NaN values
data_masked = data_masked.astype(float).dropna()
data_unmasked = data_unmasked.astype(float).dropna()

# Concatenate masked no perturbation data
no_pert_concat_masked = np.concatenate((data_masked[0:40], data_masked[80:120], data_masked[160:200]), axis=0)

# Concatenate unmasked no perturbation data
no_pert_concat_unmasked = np.concatenate((data_unmasked[0:40], data_unmasked[80:120], data_unmasked[160:200]), axis=0)

# Calculate mean and std for masked no perturbation data
mean_masked_no_pert = np.nanmean(no_pert_concat_masked, axis=0)
std_masked_no_pert = np.nanstd(no_pert_concat_masked, axis=0)

# Calculate mean and std for unmasked no perturbation data
mean_unmasked_no_pert = np.nanmean(no_pert_concat_unmasked, axis=0)
std_unmasked_no_pert = np.nanstd(no_pert_concat_unmasked, axis=0)

# Calculate mean and std for masked gradual perturbation data
mean_masked_grad = np.nanmean(data_masked[40:79], axis=0)
std_masked_grad = np.nanstd(data_masked[40:79], axis=0)

# Calculate mean and std for unmasked gradual perturbation data
mean_unmasked_grad = np.nanmean(data_unmasked[40:79], axis=0)
std_unmasked_grad = np.nanstd(data_unmasked[40:79], axis=0)

# Calculate mean and std for masked sudden perturbation data
mean_masked_sud = np.nanmean(data_masked[120:159], axis=0)
std_masked_sud = np.nanstd(data_masked[120:159], axis=0)

# Calculate mean and std for unmasked sudden perturbation data
mean_unmasked_sud = np.nanmean(data_unmasked[120:159], axis=0)
std_unmasked_sud = np.nanstd(data_unmasked[120:159], axis=0)

# Calculate mean and std for masked overt perturbation data
mean_masked_ouv = np.nanmean(data_masked[200:239], axis=0)
std_masked_ouv = np.nanstd(data_masked[200:239], axis=0)

# Calculate mean and std for unmasked overt perturbation data
mean_unmasked_ouv = np.nanmean(data_unmasked[200:239], axis=0)
std_unmasked_ouv = np.nanstd(data_unmasked[200:239], axis=0)

# Output all values in a clear table
print('Masked No Perturbation')
print('mean:', mean_masked_no_pert)
print('std:', std_masked_no_pert)

print('Unmasked No Perturbation')
print('mean:', mean_unmasked_no_pert)
print('std:', std_unmasked_no_pert)

print('Masked Gradual Perturbation')
print('mean:', mean_masked_grad)
print('std:', std_masked_grad)

print('Unmasked Gradual Perturbation')
print('mean:', mean_unmasked_grad)
print('std:', std_unmasked_grad)

print('Masked Sudden Perturbation')
print('mean:', mean_masked_sud)
print('std:', std_masked_sud)

print('Unmasked Sudden Perturbation')
print('mean:', mean_unmasked_sud)
print('std:', std_unmasked_sud)

print('Masked Overt Perturbation')
print('mean:', mean_masked_ouv)
print('std:', std_masked_ouv)

print('Unmasked Overt Perturbation')
print('mean:', mean_unmasked_ouv)
print('std:', std_unmasked_ouv)

# Plotting the means and standard deviations
conditions = ['No Perturbation', 'Gradual Perturbation', 'Sudden Perturbation', 'Overt Perturbation']
means_masked = [np.nanmean(mean_masked_no_pert), np.nanmean(mean_masked_grad), np.nanmean(mean_masked_sud), np.nanmean(mean_masked_ouv)]
# stds_masked = [np.nanstd(mean_masked_no_pert), np.nanstd(mean_masked_grad), np.nanstd(mean_masked_sud), np.nanstd(mean_masked_ouv)]
means_unmasked = [np.nanmean(mean_unmasked_no_pert), np.nanmean(mean_unmasked_grad), np.nanmean(mean_unmasked_sud), np.nanmean(mean_unmasked_ouv)]
# stds_unmasked = [np.nanstd(mean_unmasked_no_pert), np.nanstd(mean_unmasked_grad), np.nanstd(mean_unmasked_sud), np.nanstd(mean_unmasked_ouv)]

x = np.arange(len(conditions))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(12, 6))
rects1 = ax.bar(x - width/2, means_masked, width, label='Masked', capsize=5)
rects2 = ax.bar(x + width/2, means_unmasked, width, label='Unmasked', capsize=5)


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Mean Error Angle')
ax.set_title('Means and Standard Deviations by Condition')
ax.set_xticks(x)
ax.set_xticklabels(conditions)
ax.legend()

fig.tight_layout()

plt.show()
```
